Here the network is an igraph network object, every node is indexed by an integer value starting from 0

```python
G = nx.read_graphml('./ignored_datasets/dhaka_tribune_statement_network_connected_component.graphml')
```

```python
G_integer = nx.convert_node_labels_to_integers(G, label_attribute='name')
```

```python
with open('./ignored_datasets/dhaka_tribune_statement_network_connected_component_edgelist.txt', 'w') as f:
    for edge in G_integer.edges():
        if '' in edge or ' ' in edge or edge[0] == edge[1]:
            continue
        #print edge
        f.write('\t'.join(map(str,edge)))
        f.write('\n')
```

```python
names_dict = nx.get_node_attributes(G_integer, 'name')
```

```python
roles_dict = dict()
with open('./ignored_datasets/roles.txt', 'r') as f:
    print f.next()
    print f.next()
    for line in f:
        splitted_line = line.strip().split('\t')
        #print splitted_line
        roles_dict[int(splitted_line[0])] = int(splitted_line[1])
```

```python
name_to_role_dict = {name:roles_dict[node_id] for node_id, name in names_dict.items()}
```

```python
role_based_cluster_dict = defaultdict(list)
for name, role in name_to_role_dict.items():
    role_based_cluster_dict[role].append(name)
```

```python
z =  [sorted(x) for x in role_based_cluster_dict.values()]
for line in z:
    print line
    print '\n\n'
```
